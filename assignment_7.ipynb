```python
import numpy as np
import pandas as pd
```

# Reading csv file

```python
data=pd.read_csv('/content/spam (1).csv',encoding='windows=1252')
```

```python
data.head()
```

*labeling* *columns*

```python
pd.set_option('display.max_colwidth',None)
messages=data[['v1','v2']]
messages.columns=["label","text"]
messages.head()
```

# Perprocessing unsing Gensim library

```python
import gensim
messages['text_clean_gensim'] = messages['text'].apply(lambda x: gensim.utils.simple_preprocess(x,deacc=True))
messages.head()
```

```python
messages['label']=messages['label'].map({'spam':0,'ham':1})
```

# test-train-split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(messages['text_clean_gensim'],messages['label'],test_size=0.2)
```

# Word2vec





```python
w2v_model=gensim.models.Word2Vec(X_train,min_count=1)
w2v_model.wv['free']    #ndarray corresponding to free
```

```python
w2v_model.wv.most_similar('go')   #similar words
```

```python
words=w2v_model.wv.index2word  #word count
words
```

```python
X_train_vect = np.array([np.array([w2v_model.wv[i] for i in ls if i in words]) for ls in X_train])
X_test_vect = np.array([np.array([w2v_model.wv[i] for i in ls if i in words]) for ls in X_test])
```

```python
X_train_vect_avg = []
for v in X_train_vect:
    if v.size:
        X_train_vect_avg.append(v.mean(axis=0))
    else:
        X_train_vect_avg.append(np.zeros(100, dtype=float))
        
X_test_vect_avg = []
for v in X_test_vect:
    if v.size:
        X_test_vect_avg.append(v.mean(axis=0))
    else:
        X_test_vect_avg.append(np.zeros(100, dtype=float))
```

# Random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
  
clf = RandomForestClassifier(n_estimators = 100) 
clf.fit(X_train_vect_avg, y_train)
```

```python
from sklearn import metrics
y_pred = clf.predict(X_train_vect_avg) 
metrics.accuracy_score(y_pred, y_train)
```

```python
from sklearn import metrics
y_pred = clf.predict(X_test_vect_avg) 
metrics.accuracy_score(y_pred, y_test)
```

model has an accuracy of 94%
